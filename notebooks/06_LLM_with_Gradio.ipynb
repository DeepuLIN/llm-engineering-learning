```python
import os
from openai import OpenAI
```

```python
ollama_url = "http://localhost:11434/v1"
ollama = OpenAI(api_key="ollama", base_url=ollama_url)
```

```python
system_message = "You are a helpful assistant"

def message_llamma(prompt):
    messages =[
        {"role":"system", "content": system_message}, {"role":"user", "content": prompt}
        ]
    response = ollama.chat.completions.create(
        model="llama3.2", 
        messages=messages)
    return response.choices[0].message.content
```

```python
def shout(text):
    print(f"shout has benn called with input {text}")
    return text.upper()
```

```python
shout("hello")
```

```python
import gradio as gr # oh yeah
from doctest import OutputChecker


gr.Interface(fn= shout, inputs="textbox", outputs="textbox", flagging_mode="never").launch()
```

```python
message_input = gr.Textbox(label="Your message:", info="Enter a message for lamma 3.2", lines= 7)
message_output = gr.Textbox(label = "Response:", lines = 8)

view = gr.Interface(
    fn=message_llamma,
    title= "llama",
    inputs =[message_input],
    outputs = [message_output],
    examples =["hello", "howdy"],
    flagging_mode="never"
)
view.launch()
```

```python
system_message = "YOu are a helpful assistant that responds in markdown not in code blocks"

message_input = gr.Textbox(label = "Your message", info = "Enter a message for locally run lamma 3,2", lines = 7)
message_output = gr.Markdown(label= "Response:")
view = gr.Interface(
    fn=message_llamma,
    title="llama",
    inputs=[message_input],
    outputs=[message_output],
    examples=[
        "Explain the transformer architecture to a laymen"
        ],
    flagging_mode="never"
    )
view.launch()
```

```python
def stream_llama(prompt):
    messages = [
        {"role": "system", "content": system_message},
        {"role": "user", "content": prompt}
      ]
    stream = ollama.chat.completions.create(
        model='llama3.2',
        messages=messages,
        stream=True
    )
    result = ""
    for chunk in stream:
        result += chunk.choices[0].delta.content or ""
        yield result
```

```python
message_input = gr.Textbox(label = "Your message", info = "Enter a message for locally run lamma 3,2", lines = 7)
message_output = gr.Markdown(label= "Response:")
view = gr.Interface(
    fn=stream_llama,
    title="llama",
    inputs=[message_input],
    outputs=[message_output],
    examples=[
        "Explain the transformer architecture to a laymen"
        ],
    flagging_mode="never"
    )
view.launch()
```

```python
def stream_phi3(prompt):
    messages = [
        {"role": "system", "content": system_message},
        {"role": "user", "content": prompt}
      ]
    stream = ollama.chat.completions.create(
        model='llama3.2',
        messages=messages,
        stream=True
    )
    result = ""
    for chunk in stream:
        result += chunk.choices[0].delta.content or ""
        yield result
```

```python
message_input = gr.Textbox(label = "Your message", info = "Enter a message for locally run phi3", lines = 7)
message_output = gr.Markdown(label= "Response:")
view = gr.Interface(
    fn=stream_phi3,
    title="phi3:mini ",
    inputs=[message_input],
    outputs=[message_output],
    examples=[
        "Explain the transformer architecture to a laymen"
        ],
    flagging_mode="never"
    )
view.launch()
```

```python
def stream_model(prompt, model):
    if model=="llama3.2":
        result = stream_llama(prompt)
    elif model=="phi3:mini":
        result = stream_phi3(prompt)
    else:
        raise ValueError("Unknown model")
    yield from result
```

```python
message_input = gr.Textbox(label="Your message:", info="Enter a message for the LLM", lines=7)
model_selector = gr.Dropdown(["llama3.2", "phi3:mini"], label="Select model", value="phi3:mini")
message_output = gr.Markdown(label="Response:")

view = gr.Interface(
    fn=stream_model,
    title="LLMs", 
    inputs=[message_input, model_selector], 
    outputs=[message_output], 
    examples=[
            ["Explain the Transformer architecture to a layperson", "llama"],
            ["Explain the Transformer architecture to an aspiring AI engineer", "PHi3"]
        ], 
    flagging_mode="never"
    )
view.launch()
```
