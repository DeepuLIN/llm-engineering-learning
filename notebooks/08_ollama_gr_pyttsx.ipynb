```python
import gradio as gr
import requests
import pyttsx3
```

```python
def ask_ollama(prompt):
    url = "http://localhost:11434/api/generate"

    response = requests.post(url, json={
        "model": "llama3.2:latest",
        "prompt": f"You are a flight assistant. Answer clearly.\nUser: {prompt}",
        "stream": False
    })

    return response.json()["response"]
```

```python
engine = pyttsx3.init()

def speak(text):
    engine.say(text)
    engine.runAndWait()
```

```python
def flight_assistant(user_input):
    response = ask_ollama(user_input)
    speak(response)
    return response
```

```python
demo = gr.Interface(
    fn=flight_assistant,
    inputs=gr.Textbox(label="Ask about flights "),
    outputs=gr.Textbox(label="Assistant Response"),
    title="AI Flight Assistant with Voice"
)

demo.launch()
```
